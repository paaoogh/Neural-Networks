En la página de Kaggle nos piden evaluar con RMSPE: Root Mean Squared Percentage Error.

$$RMSPE(y',y)=\sqrt{\frac{1}{n}\sum \left(\frac{y-y'}{y}\right)^2}$$
 
Donde *y'* es nuestra predicción. Ojo que donde *y=0* se ignora.

**Nota al pie: podríamos usar algo mejor para medirlo** porque esa métrica hace que siempre te convenga adivinar que hay ventas, pero también es necesario saber cuándo no habrá ventas. MSE o LU estarían bien. Pero hay que ver qué pasa si pones más/menos productos de los que se venden para saber cuánto dinero pierdes y tratar de minimizar eso.  

```python
def rmspe(yp,y):
    u = (y-yp)/y
    return torch.sqrt((u*u).mean())
```

```python
df = df[df['Sales']>0].copy() #porque ignoramos 0

df['Sales'] = df['Sales'].astype('float64')
#Ojo porque Sales tiene números grandores. Entonces habrá NaN
```

```python
df['Sales'] /= 1000.
#El dividir o multiplicar por constante no cambia lo de la suma
```

Dividir entre variables categóricas, contínuas y "y": Customers de hecho, en el concurso, ni siquiera está incluido como un posible valor para entrenar. 

```python
cat_names = ['Store', 'DayOfWeek', "Open", "Promo", "SchoolHoliday", "StateHoliday"]
cont_names = [] #Ahora mismo no hay variables continuas
y_names = ['Sales']
```

```python
len(df)
```

```python
X = list(range(len(df)))
```

```python
valid_cut = len(df) - 20000 #dejo los últimos como validación
```

```python
#Para crear un databunch 
src = ft.TabularPandas(df,
                       procs=[ft.Categorify, ft.FillMissing], #esto es como equivalente a las transforms
                       #Categorify agarra como diccionario de categorías a números
                       #Fill missing reemplaza con la media
                       #evita normalizar, igual ya tienes batch_norm
                       cat_names =cat_names, cont_names=cont_names,y_names=y_names,
                       splits=(X[:valid_cut],X[valid_cut:]))
```

En los datos tabulares podemos usar números muchísimo más grandes que en las imágenes para los batchsizes porque para las imágenes tenemos cosas de (algo x algo) y son más números que ahora. 

```python
dls = src.dataloaders(bs=2048)
```

```python
learn = ft.tabular_learner(dls, opt_func=ft.ranger, metrics=rmspe)
```

## Mejoras

### Función de pérdida. 

Como dijimos arriba, **no estamos optimizando lo que queremos optimizar**, estamos más bien usando MSE pero nos pidieron MSPE (el sqrt no importa). Es de hecho, diferenciable.

```python
def mspe(yp,y):
    u = (y-yp)/y
    return (u*u).mean()
```

```python
learn = ft.tabular_learner(dls,
                           loss_func = mspe,
                           opt_func  = ft.ranger,
                           metrics = rmspe)
```

```python
learn.fit_one_cycle(20, 1e-3,div=2,pct_start=0.5)
```

## Modelo

Lo siguiente que dijimos que podía mejorar era el modelo... porque cambiar la función de pérdida no ayudó mucho realmente. 

Fastai nos permite configurar el modelo con **layers** (una lista con las capas intermedias), ¿recuerdas? Además, podemos ajustar los tamaños de los embeddings con un diccionario llamado **emb_szs**. Con la variable **config** también podemos configurar cosas de un modelo ya creado. 

Claro, podríamos crear un modelo propio, pero fastai ya lo hace muy bien; de entrada, no hay que preocuparnos por el tamaño de los embeddings.

```python
learn = ft.tabular_learner(dls,
                           loss_func = mspe,
                           #embeddings sizes = cambiar el tamaño de los embeddings
                           emb_szs = {'Store':16,},
                           #layers espera una lista con números: el número de neuronas en cada capa
                           layers = [256, 128, 128],
                           #config: aquí puedo poner más cosas
                           config = ft.tabular_config(act_cls=nn.LeakyReLU(inplace=True)),
                           opt_func = ft.ranger,
                           metrics=rmspe)
```

```python
learn.model
#veamos que sí cambió las activaciones y agregó capas ocultas
```

```python
learn.fit_one_cycle(20, 1e-2,div=2,pct_start=0.5)
```

Aún con eso vemos que no ha mejorado mucho. De hecho, es muy mínimo.

## Agregar datos

Para empezar, tenemos datos sobre la fecha que no hemos usado para NA-DA. ¿Qué tal que hay meses que venden más que otros?

```python
df = pd.read_csv("Data/train.csv", parse_dates=['Date'])
```

```python
df = df.sort_values(['Date']).copy()
ft.add_datepart(df,'Date') 
#Esta última función toma un data frame y dónde está la fecha
#añade columnas que son relevantes a la fecha
#Inicio del mes, fin de año, etc. 
```

```python
df.drop('DayOfWeek', axis=1,inplace=True) #porque lunes vs domingo
```

```python
df = df[df['Sales']>0].copy()

df['Sales'] = df['Sales'].astype('float64')/1000 
```

Ojo: tenemos datos sobre las tiendas: qué tipos de tiendas y así. La red neuronal podría usarlas. La distancia a la copetencia, etc... 

```python
stores = pd.read_csv("Data/store.csv")
```

```python
stores
```

```python
df_stores = df.join(stores.set_index('Store'),on='Store')
df_stores
```

```python
df_stores.head().T #para ver todas las columnas
```

```python
cat_names = ['Store',"Dayofweek","Open",'Promo','StateHoliday', 'SchoolHoliday',
            'Year','Month','Week','Dayofweek',
            'Is_month_end','Is_month_start','Is_quarter_end','Is_quarter_start','Is_year_end',
             'Is_year_start','StoreType','Assortment','CompetitionOpenSinceYear','CompetitionOpenSinceMonth',
            'Promo2','PromoInterval']

cont_names = ['Day','Dayofyear','Elapsed','CompetitionDistance', 'CompetitionOpenSinceMonth', 'Promo2SinceWeek',
             'Promo2SinceYear']

y_names = ['Sales']
```

```python
X = list(range(len(df)))
```

```python
valid_cut = len(df)-20000
```

```python
src = ft.TabularPandas(df_stores,
                       procs=[ft.Categorify, ft.FillMissing], 
                       cat_names =cat_names, cont_names=cont_names,y_names=y_names,
                       splits=(X[:valid_cut],X[valid_cut:]))
```

```python
dls = src.dataloaders(bs=2048)
```

```python
learn = ft.tabular_learner(dls,
                           loss_func = mspe,
                           emb_szs = {'Stores':16},
                           layers = [256,128],
                           config = ft.tabular_config(act_cls=nn.LeakyReLU(inplace=True)),
                           opt_func = ft.ranger,
                           metrics=rmspe)
```

```python
learn.model
```

```python
learn.fit_one_cycle(20,1e-2,div=2,pct_start=0.5)
```
